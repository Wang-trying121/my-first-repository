```python
import pandas as pd
# 读取Excel文件
df = pd.read_excel('学生成绩数据.xlsx')#, usecols = ['姓名','语文','数学','英语','物理','化学','生物'])
#数据处理
df['总成绩'] = df['语文'] + df['数学'] + df['英语'] + df['物理'] + df['化学'] + df['生物']
#写入数据
df.to_excel('学生成绩数据_01.xlsx', index=False)
#检查
print(df['总成绩'].max())#最高分
print(df['总成绩'].min())#最低分
print(df['总成绩'].mean())#平均分
```

```python

#数据查看—————— head tail describe info shape columns
print(df.head(10))#显示前10行数据
print(df.tail(10))#显示后10行数据
# print(df.describe())#数据统计描述  std-标准差
print(df.info())#数据信息
print(df.shape)#数据行列(1000, 11)  行列数
print(df.columns)#数据列名 #Index(['学号', '姓名', '语文', '数学', '英语', '物理', '化学', '生物', '体育', '美术', '音乐'], dtype='str')



```

```python
#选择数据————
#选择列
#单列
print(df['姓名'])#单列数据
#多列
print(df[['姓名', '语文', '数学']])#多列数据
#选择行————iloc loc[start:stop:step]
print(df.iloc[0:10])#显示前10行数据 基于行号选择 左闭右开
print(df.loc[0:10])#显示前10行数据 基于索引选择 全闭
```

```python
#数据过滤
#数字筛选
df[df['语文'] >= 150]
df[df['数学'].between(60,90)] #默认全闭
#文字筛选
df[df['体育'].isin(['A'])]
#逻辑关系
#和
df[(df['语文'] >= 150) & (df['数学'] >= 90)]
#或
df[(df['语文'] >= 150) | (df['数学'] >= 90)]
 #非
df[~(df['语文'] >= 150)]
```

```python
#数据清洗
#处理缺失值
#查看缺失值
df.isnull()
#删除缺失值
df.dropna()#删除行
df.dropna(axis=1)#删除列
#填充缺失值
df.fillna(0)#用0填充
df.ffill()#用前一个值填充
df.bfill()#用后一个值填充
df.fillna(value={'语文': 0, '数学': 0, '英语': 0, '物理': 0, '化学': 0, '生物': 0})#用指定值填充

```

```python
#重复值处理
#查看重复值
df.duplicated()#判断所有列数据全都一样
df.duplicated(subset=['姓名'])#判断指定列数据一样
#删除重复值
df.drop_duplicates(subset = ['姓名'], keep='first')#删除重复值，只保留第一个
df.drop_duplicates(subset = ['姓名'], keep='last')#删除重复值，只保留最后一个
df.drop_duplicates(subset = ['姓名'], keep=False)#删除重复值，不留任何

```

```python
#处理异常值
#查看异常值
df[df['语文'] > 150]#查看异常值
#删除异常值
df.drop(df[df['语文'] > 150].index)
#修复异常值
df['语文'].replace(150, 100)#修复异常值
```
